# fastQC

Need the total number of sequenced reads of good quality

````python
%time df_fastqc = pd.read_csv('./metadata/RNAseq_FastQC.csv.gz', compression='gzip')
````

````python
df_fastqc.head(1)
````

````python
df_fastqc = df_fastqc[df_fastqc.CGHub_analysisID.isin(df_gc_pivot.analysis_id)]
````

````python
def process_fastqc_record(grp):
    assert (grp.minLength == grp.maxLength).all()
    assert grp.minLength.unique().shape[0] == 1
    read_len = grp.minLength.unique()[0]
    num = (grp.numTotalSeq - grp.numPoorSeq).sum()
    return pd.Series({'numGoodSeq': num, 'read_length': read_len})
````

````python
# proc: processed
%time df_fastqc_proc = df_fastqc.groupby('CGHub_analysisID').apply(process_fastqc_record)
````

````python
df_fastqc_proc.shape
````

````python
df_fastqc_proc.head(1)
````

````python
df_gc_pivot.head(1)
````

````python
merged = df_gc_pivot.merge(df_fastqc_proc, left_on='analysis_id', right_index=True)
````

# Convert coverage to RPKMS

adopted from http://www.metagenomics.wiki/pdf/definition/rpkm-calculation:
```
RPKMS =   numReads / ( geneLength/1000 * totalNumReadsSequened/1,000,000 )
```

````python
df_rpkms = merged.copy()
for col in TARGET_GENES:
    df_rpkms[col] = (merged[col] * 1e9) / (merged.read_length * merged.numGoodSeq)
````

````python
df_rpkms.describe()
````

````python
df_rpkms.to_csv('./results/RPKMS.csv', index=False)
````
